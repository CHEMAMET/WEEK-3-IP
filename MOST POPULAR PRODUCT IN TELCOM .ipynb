DATA CLEANING

```python
#since all the three datasets have similar columns, they can be combined for easier clean up and analysis

df_data = pd.concat([df_1,df_2,df_3], axis=0)
df_data.info()
```

```python
#For ease of calculation, change the data types of data columns which store date values to date 

df_data['Date_Time'] = pd.to_datetime(df_data['Date_Time'])
#check if the data type changed
df_data.info()
```

```python
#to git rid of those columns that are not required, we drop them
df_data.drop(['Country_A', 'Country_B'], axis=1)
```

```python
#drop duplicates as well
df_data.drop_duplicates()
```

```python
#review changes
df_data.info()
```

```python
#merge the Telcom dataset and the cells_geo dataset
df_data_geo = pd.merge(df_data,df_geo, how='inner', on=['Cell_ID'])

df_data_geo.info()
```
